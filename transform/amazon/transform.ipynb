# Importing csv files

```python
base_path = '/content/drive/MyDrive/Book Space/Desenvolvimento/ETL/extracoes/'
```

```python
file_names = []
for root, dirs, files in os.walk(base_path):
  for file in files:
    file_names.append(file)
print(file_names)
```

```python
file_paths = []
for file in file_names[:len(file_names) - 2]:
    file_paths.append(base_path+'/'+file)
print(file_paths)
```

```python
df = pd.DataFrame()
for file_path in file_paths:
    new_data = pd.read_csv(file_path)
    df = pd.concat([df, new_data], axis=0)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
display(df)
```

# Data pre-processing

* Alterar coluna 'paginas' para int
* Corrigir erro de informação na coluna 'editora'

```python
df['genero'].value_counts()
```

## Column 'page'

```python
new_page_data = []
for pagina in df['paginas']:
    new_page_data.append(int(pagina.replace(' páginas', '')))
df['paginas'] = new_page_data
df['paginas']
```

## Column 'editora'

```python
df['editora']
```

```python
editoras = []
for editora in df['editora']:
  if ";" in editora:
    index = editora.index(";")
    nova_editora = editora[:index]
  elif " (" in editora:
    index = editora.index(" (")
    nova_editora = editora[:index]
  editoras.append(nova_editora)
print(df['editora'].shape[0], len(editoras))
```

```python
df['editora'] = editoras
df['editora'].value_counts()
```

## Concat with new gender data

```python
gender_path =  base_path+'generos/'+file_names[-1]
print(gender_path)
```

```python
gender_data = pd.read_csv(gender_path, sep=",", names=["titulo", 'generos'])
gender_data.drop(0, axis=0, inplace=True)
gender_data.head(15)
```

```python
len(df['titulo']), len(gender_data['titulo'])
```

```python
gender_data = gender_data.sort_values(by='titulo').reset_index(drop=True)
gender_data.info()
```

```python
df = df.sort_values(by='titulo').reset_index(drop=True)
df.info()
```

```python
def remove_book_from_df():
  index = 0
  for titulo in gender_data['titulo']:
      # print(str(index) + '. ' + titulo + ' --- ' + df.iloc[index].titulo)
      if titulo not in df.iloc[index].titulo:
        df.drop(index, axis=0, inplace=True)
        break
      index = index + 1
```

```python
while df.shape[0] != gender_data.shape[0]:
  remove_book_from_df()
```

```python
df.head()
```

```python
df = pd.merge(df, gender_data, left_index=True, right_index=True)
df
```

```python
df.drop(['genero', 'titulo_y'], axis=1, inplace=True)
```

```python
df.columns = ['titulo', 'descricao', 'paginas', 'idioma', 'editora', 'autor',
       'imgLink', 'isbn13', 'isbn10', 'amazonLink', 'generos']
df = df[['titulo', 'descricao', 'paginas', 'idioma', 'editora', 'autor',
       'generos', 'imgLink', 'isbn13', 'isbn10', 'amazonLink']]
```
